```python
import pandas as pd
import numpy as np
import pickle
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.grid_search import GridSearchCV
```

```python
import os
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
os.chdir('../output_data')
```

```python
df_lr = pd.read_csv('train_lr.csv',header=0)
```

```python
os.chdir('../Models')
```

```python
x = df_lr.drop(['Close'],axis=1)
```

```python
y = np.array(df_lr['Close'])
```

```python
tuned_parameters={'n_estimators':[100*i for i in range(20,26)]}
model = GridSearchCV(GradientBoostingRegressor(learning_rate=.1,min_samples_leaf=.01,max_depth=10,subsample=.8,
                                               random_state=42),tuned_parameters,cv=5,scoring='r2',verbose=50)
```

```python
model.fit(x,y)
```

```python
model.best_estimator_
```

```python
model.best_score_
```
